Encoding 1 day candles for 5 days and doing rag - as a test

```python
import yfinance as yf
import pandas as pd
```

```python
dat = yf.Ticker("SPY")
data = dat.history( period="10y", interval="1d")
```

```python
dat.info
```

```python
avgVol = dat.info['averageVolume']

def applyVolumeNormalizedVectorized(row):
    row['VolumeNormalized'] = (row['Volume'] - avgVol) / avgVol
    return row


data = data.apply(applyVolumeNormalizedVectorized, axis=1)
data
```

```python
avgMovement = (data['Close']-data['Open']).mean()


def applyMovementNormalized(row):
    row['MovemementNormalize'] = (row['Close'] - row['Open']) / avgMovement
    return row


data = data.apply(applyMovementNormalized, axis=1)
data
```

```python
megaDb = pd.DataFrame(columns=['vol1', 'mov1', 'vol2', 'mov2', 'vol3', 'mov3', 'vol4', 'mov4', 'vol5', 'mov5'])
 # new db for the vectors


sliding_window = 5
cleanDB = data[['VolumeNormalized', 'MovemementNormalize']]
#cleanDB= cleanDB[0:10]
#cleanDB



for i in range(len(cleanDB) - (sliding_window-1)):
    ar = []

    for j in range(sliding_window):

        ar.append(cleanDB.iloc[i+j].VolumeNormalized)
        ar.append(cleanDB.iloc[i+j].MovemementNormalize)


    megaDb.loc[i] = ar
    #print(ar)

megaDb
```
